### In this whole workshop we are going to use the IRIS dataset. Here is the link for downloading : https://archive.ics.uci.edu/dataset/53/iris

```python
import pandas as pd
```

#### You can either download the dataset and follow these steps to read the csv file

```python
df = pd.read_csv("iris.data")
```

#### Or if you have a url to the dataset like this: https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data , then you should follow these steps:

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df = pd.read_csv(url, header=None, names=columns)
```

##### Now we will look into the basic properties like how many rows and columns are present in the dataset

```python
df.describe()
```

```python
print("\nBasic Data Properties:")
print("Number of Rows and Columns:")
print(df.shape)
```

##### To get a clear picture of dataset like which feature variable are there, we check for the column names of the dataset.

```python
print("\nColumn Names:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nFiltering Data:")
setosa_df = df[df['species'] == 'Iris-setosa']
print("Setosa Species Data:")
print(df['species'])
```

```python
print("\nSorting Data:")
sorted_df = df.sort_values(by='sepal_length', ascending=True)
print("Sorted Data by Sepal Length:")
print(sorted_df.head())
```

```python
missing_data = df.isnull().sum()
print("\nMissing Data:")
print(missing_data)
```

#### Calculate statistical data:

```python
mean = df['sepal_length'].mean()
print(mean)
median = df['sepal_length'].median()
print(median)
std_deviation = df['sepal_length'].std()
print(std_deviation)
variance = df['sepal_length'].var()
print(variance)
total = df['sepal_length'].sum()
print(total)
count = df['sepal_length'].count() #Count the number of non-null values in a column using the count() method.
print(count)
min_value = df['sepal_length'].min()
print(min_value)
max_value = df['sepal_length'].max()
print(max_value)
# correlation_matrix = df.corr()
# covariance_matrix = df.cov()
# print(correlation_matrix, covariance_matrix)
quantile_25 = df['sepal_length'].quantile(0.25)
quantile_50 = df['sepal_length'].quantile(0.50)
quantile_75 = df['sepal_length'].quantile(0.75)
print(quantile_25, quantile_50, quantile_75)
value_counts = df['species'].value_counts() #Count the unique values in a column and their frequencies using the value_counts() method.
print(value_counts)
```

# Numpy:It is a library for numerical computations in python

```python
import numpy as np
```

### First of all we need to convert the dataframe into a numpy array, so here it is

```python
# Convert the Iris dataset into NumPy arrays
iris_array = df.to_numpy()
```

#### Now display the 1st 5 rows in array, here we cannot use the function head() that we used for pandas. We applying indexing and slicing operations for Numpy arrays.

```python
# Display the first few rows of the NumPy array
print("\nIris Data as NumPy Array:")
print(iris_array[:5])
```

#### Now, let us see some basic operations which we can perform using numpy

```python
print("\nBasic Array Operations:")
# Calculate the mean of sepal lengths
mean_sepal_length = np.mean(iris_array[:, 0])
print("Mean Sepal Length:", mean_sepal_length)

# Find the maximum sepal width
max_sepal_width = np.max(iris_array[:, 1])
print("Maximum Sepal Width:", max_sepal_width)

# Find the minimum petal length
min_petal_length = np.min(iris_array[:, 2])
print("Minimum Petal Length:", min_petal_length)
```

#### Broadcasting with Numpy

```python
# Demonstrate broadcasting with NumPy
print("\nBroadcasting with NumPy:")
# Increase all sepal lengths by 1
iris_array[:, 0] += 1
print("Updated Sepal Lengths:")
print(iris_array[:, 0])
print(len(iris_array))
```

#### Generating random data with Numpy

```python
# Generate random data using NumPy functions
print("\nGenerating Random Data with NumPy:")
# Generate an array of random integers between 1 and 100
random_data = np.random.randint(1, 100, size=(5, 3))
print("Random Data:")
print(random_data)
```

# Matplotlib:

```python
import matplotlib.pyplot as plt

# Create simple scatterplots to visualize feature comparisons
plt.figure(figsize=(10, 1))

# Scatterplot for sepal length vs. sepal width
plt.scatter(df['sepal_length'], df['sepal_width'], c='red', label='Sepal')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Scatterplot for petal length vs. petal width
plt.scatter(df['petal_length'], df['petal_width'], c='blue', label='Petal')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

plt.title('Sepal and Petal Comparisons')
plt.legend()
plt.show()

# Create multiple plots and subplots for different visualizations
plt.figure(figsize=(12, 6))

# Scatterplot for sepal length vs. sepal width
plt.subplot(1, 2, 1)
plt.scatter(df['sepal_length'], df['sepal_width'], c='green')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')

# Scatterplot for petal length vs. petal width
plt.subplot(1, 2, 2)
plt.scatter(df['petal_length'], df['petal_width'], c='purple')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs. Petal Width')

plt.tight_layout()
plt.show()

# Save the plots to files for later use
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal_length'], df['sepal_width'], c='orange')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.savefig('sepal_comparison.png')

plt.figure(figsize=(8, 6))
plt.scatter(df['petal_length'], df['petal_width'], c='pink')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs. Petal Width')
plt.savefig('petal_comparison.png')

plt.tight_layout()
plt.show()
```

#### Some more interesting things to do with matplotlib:


```python
# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Create scatterplots for feature comparisons
# Scatterplot for sepal length vs. sepal width
axes[0, 0].scatter(df['sepal_length'], df['sepal_width'], c='blue', label='Sepal')
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Sepal Width (cm)') #plots the subplot in left-top region of the grid axes[0,0]

# Scatterplot for petal length vs. petal width
axes[0, 1].scatter(df['petal_length'], df['petal_width'], c='green', label='Petal')
axes[0, 1].set_xlabel('Petal Length (cm)')
axes[0, 1].set_ylabel('Petal Width (cm)') #plots the subplot in right-top region of the grid axes[0,1]

# Customize the plots with labels, titles, and colors
for ax in axes.flat:
    ax.set_title('Iris Dataset')
    ax.legend()

# Create a histogram for sepal length
axes[1, 0].hist(df['sepal_length'], bins=15, color='purple')
axes[1, 0].set_xlabel('Sepal Length (cm)')
axes[1, 0].set_ylabel('Frequency') #plots the subplot in left-bottom region of the grid axes[1,0]


# Create a histogram for petal length
axes[1, 1].hist(df['petal_length'], bins=15, color='orange')
axes[1, 1].set_xlabel('Petal Length (cm)')
axes[1, 1].set_ylabel('Frequency') #plots the subplot in right-bottom region of the grid axes[1,1]

# Adjust layout and spacing
plt.tight_layout()

# Save the plots to files for later use
plt.savefig('iris_scatterplots.png')
plt.savefig('iris_histograms.png')

# Show the plots
plt.show()
```

# Seaborns:

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Use Seaborn to create advanced plots for the Iris dataset
sns.set(style="whitegrid")

# Pair plot for visualizing relationships between features
sns.pairplot(df, hue="species", markers=["o", "s", "D"], height=3)
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()

# Violin plot to visualize feature distributions
plt.figure(figsize=(8, 6))
sns.violinplot(x="species", y="sepal_length", data=df, palette="pastel")
plt.title("Violin Plot of Sepal Length by Species")
plt.show()

# Box plot to visualize feature distributions
plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="petal_width", data=df, palette="pastel")
plt.title("Box Plot of Petal Width by Species")
plt.show()

# Customize Seaborn plots by adjusting styles and colors (Seaborn styles are already set)

# Combine Seaborn plots with Pandas DataFrames and Matplotlib for comprehensive data analysis
# For example, create a Pandas DataFrame plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df, palette="dark")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs. Sepal Width by Species")
plt.show()
```
